### code to clean the dataset

```python
import re
def cleanResume(txt):
    cleanText = re.sub('http\S+\s', ' ', txt)
    cleanText = re.sub('RT|cc', ' ', cleanText)
    cleanText = re.sub('#\S+\s', ' ', cleanText)
    cleanText = re.sub('@\S+', '  ', cleanText)  
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) 
    cleanText = re.sub('\s+', ' ', cleanText)
    return cleanText
```

```python
cleanResume("hellwo world and the with : , : '  end https:\\facebook.com vivek choudyhr @gmai.com vc")
```

```python
df["Resume"] = df["Resume"].apply(lambda x: cleanResume(x))
```

### word into catogorical data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lc  = LabelEncoder()
```

```python
lc.fit(df['Category'])
df["Category"] = lc.transform(df["Category"])
```

### vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words="english")

tfidf.fit(df["Resume"])
output = tfidf.transform(df["Resume"])
```

```python
output
```

```python
df["Resume"][0]
```

### test train split


```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(output, df["Category"], test_size=0.33, random_state=42)
```

```python
xtrain.shape
```

```python
xtest.shape
```

### model tranining

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(xtrain, ytrain)
ypred = clf.predict(xtest)
print(accuracy_score(ytest, ypred))
```
